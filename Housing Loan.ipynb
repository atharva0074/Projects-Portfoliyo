Step 1 - Gathering the Data

```python
#import packages
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/Housing.csv")
```

```python
data.head()
```

Step 2 - Data Cleaning

```python
data['mainroad']=data['mainroad'].apply({'yes':1,'no':0}.get)
data['guestroom']=data['guestroom'].apply({'yes':1,'no':0}.get)
data['basement']=data['basement'].apply({'yes':1,'no':0}.get)
data['hotwaterheating']=data['hotwaterheating'].apply({'yes':1,'no':0}.get)
data['airconditioning'] = data['airconditioning'].apply(lambda x: {'yes':1,'no':0}.get(x))
data['prefarea'] = data['prefarea'].apply(lambda x: {'yes':1,'no':0}.get(x))
data['furnishingstatus'] = data['furnishingstatus'].apply(lambda x: ['furnished', 'semi-furnished', 'unfurnished'].index(x) + 1)
```

```python
data.head()
```

Step 3 - Dividing the Data into dependent and independent varaibles

```python
x=data[['area','bedrooms','bathrooms','stories','mainroad','guestroom','basement','hotwaterheating','airconditioning','parking','prefarea','furnishingstatus']]
y=data['price']
```

Step 4 - Spliting the dataset in training and testing set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.85)
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
y_pred=regression.predict(x_test)
```

```python
print(y_pred)
```

```python
regression.score(x,y)
```

Ridge regression

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=1)
ridge.fit(x_train,y_train)
```

```python
ridge_predictions=ridge.predict(x_test)
```
